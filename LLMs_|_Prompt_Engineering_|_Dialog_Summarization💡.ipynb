## **What are LLMs?**

LLMs stands for **"Large Language Models."** These are computer programs or **artificial intelligence systems** designed to **understand and generate human language**. They are a type of **machine learning model** that has **gained significant attention and popularity** in recent years due to their **remarkable language processing capabilities**.

Here are some key points to understand about LLMs:

1. **Language Understanding**: LLMs are designed to understand and work with human languages, such as English, Spanish, or any other natural language. They can process text, analyze its meaning, generate coherent sentences, and even perform language-related tasks like translation, summarization, and question answering.

2. **Learning from Data**: LLMs are typically built using a form of machine learning called "pre-training." During pre-training, these models are exposed to massive amounts of text data from the internet. They learn patterns, grammar, vocabulary, and even some level of common-sense reasoning from this data. This phase helps them acquire a broad understanding of language.

3. **Fine-Tuning**: After pre-training, LLMs can be fine-tuned on specific tasks. This involves exposing the model to task-specific data and examples. For example, fine-tuning might involve training an LLM to generate human-like responses in a customer support chatbot or to classify news articles into categories.

4. **Generative and Discriminative**: LLMs can be both generative and discriminative. Generative LLMs, like GPT-3, are capable of generating text that is contextually relevant and coherent, often used for tasks like creative writing or content generation. Discriminative LLMs, like BERT, excel at understanding and classifying text, often used for tasks like sentiment analysis or document classification.

5. **Scalability**: One of the defining features of LLMs is their scalability. They can be made larger by increasing the number of parameters (neurons), which tends to improve their performance. However, this also requires significant computational resources.

6. **Applications**: LLMs have found applications in various domains, including natural language understanding, chatbots, language translation, content generation, recommendation systems, and much more. They have significantly advanced the field of Natural Language Processing (NLP) and have enabled machines to interact with humans in more sophisticated ways.

In summary, LLMs, or Large Language Models, are advanced machine learning models that have been trained on vast amounts of text data and are capable of understanding and generating human language. They are a foundational technology behind many of the language-related AI applications we encounter in our daily lives.

## **What is Prompt Engineering?**

**Prompt Engineering** is a technique used in the **context of Large Language Models (LLMs)** like **GPT-3, BERT, or similar models** to guide and shape their responses for **specific tasks** or **queries**. It involves designing or crafting **well-structured prompts or input instructions** to elicit **desired outputs** or responses from these **language models**. Prompt Engineering plays a crucial role in **harnessing the capabilities of LLMs** for various natural language processing tasks. Here are some key aspects of **Prompt Engineering**:

1. **Task Guidance**: LLMs are highly versatile and can perform a wide range of tasks, from text generation to language translation to question answering. Prompt Engineering helps narrow down the focus of the model by providing a task-specific prompt that directs it to perform a particular task. For example, if you want to use an LLM for translation, you can design a prompt like "Translate the following English text into French: 'Hello, how are you?'" The model will then understand the task and provide a French translation.

2. **Format and Structure**: The way prompts are structured can significantly impact the quality of the model's responses. Engineers and researchers use Prompt Engineering to experiment with different prompt formats and structures to find the most effective ones. This might include specifying input formats, asking questions in a certain way, or providing context to the model.

3. **Few-Shot Learning**: Prompt Engineering is particularly valuable for few-shot learning scenarios. In these situations, the model is given only a small number of examples or context to learn from. Engineers craft prompts that encapsulate these examples effectively to ensure the model can generalize from them and perform the task accurately. For example, in a one-shot learning scenario, a prompt might be designed to instruct the model to generate text similar to a provided example.

4. **Control and Safety**: Prompt Engineering can also be used to control the behavior of LLMs and ensure they generate safe, unbiased, or contextually appropriate responses. Engineers can add constraints, guidelines, or context within prompts to guide the model's behavior and reduce the likelihood of generating harmful or inappropriate content.

5. **Iterative Process**: Prompt Engineering often involves an iterative process of experimentation and refinement. Engineers may try various prompts, observe the model's responses, and make adjustments to improve the results.

6. **Fine-Tuning**: In some cases, fine-tuning may complement prompt engineering. Fine-tuning involves training an LLM on specific data or tasks to enhance its performance. Prompt engineering can be used in conjunction with fine-tuning to fine-tune the model's behavior in alignment with the desired task.

Overall, **Prompt Engineering** is a **creative and strategic** approach to **leverage the power of LLMs** for **specific applications** and tasks by carefully crafting **input instructions** that guide the model to **generate the desired outputs** or **responses**. It is a **critical skill** in the **field of natural language processing and AI**, enabling the **customization and control of LLMs** for various **real-world applications**.

## **What is Few Shot Learning?**

**Few-Shot Learning** is a **machine learning paradigm** that focuses on **training models** to make **accurate predictions** or perform tasks when they are given **very limited examples or data points**. In traditional **machine learning**, models often require a **substantial amount of training data** to **generalize effectively** to new, unseen examples. However, **Few-Shot Learning** aims to bridge this gap by **enabling models** to learn from a few, or even just one, training examples.

Key characteristics and concepts of Few-Shot Learning include:

1. **Limited Data**: The defining feature of Few-Shot Learning is the scarcity of training data. Instead of having thousands or millions of examples for each class or category, these models are trained with only a handful of examples per class. This makes Few-Shot Learning particularly useful in scenarios where acquiring a large dataset is challenging or impractical.

2. **N-Shot Learning**: Few-Shot Learning is often categorized based on the number of examples provided during training. Common categories include:
  * **Zero-Shot Learning**: The model has no specific examples during training for a particular class or task but is expected to recognize or generate instances of that class during testing.
  
  * **One-Shot Learning**: The model is trained with just one example per class. It must then identify or generate instances of these classes during testing.

  * **Few-Shot Learning**: The model is trained with a small number of examples (more than one but still relatively few) per class. It is expected to generalize to new examples during testing.

3. **Transfer Learning**: Few-Shot Learning often leverages transfer learning, where a pre-trained model (often a large language model like GPT-3 or a convolutional neural network like ResNet) is fine-tuned on the limited training data. This pre-trained knowledge helps the model adapt quickly to new tasks with minimal data.

4. **Feature Extraction**: In some cases, Few-Shot Learning techniques involve extracting useful features from the limited training data to aid in classification or generation tasks. These features can then be used by the model to make predictions or generate content.

5. **Applications**: Few-Shot Learning has numerous practical applications, including image classification, object recognition, natural language understanding, and language generation. For example, a Few-Shot Learning model could be trained to recognize rare species of birds with just a few images per species.

6. **Challenges**: While Few-Shot Learning is a powerful concept, it also presents challenges. Models must effectively generalize from limited data, which can lead to overfitting or suboptimal performance. Developing effective training strategies, data augmentation techniques, and appropriate evaluation metrics are ongoing areas of research.

Overall, Few-Shot Learning addresses the real-world problem of learning from minimal data, making it a valuable approach in machine learning and artificial intelligence. It enables models to adapt to new tasks and categories with limited examples, which is especially beneficial in scenarios where collecting extensive training data is costly or impractical.

## **What is Dialog Summarization?**

**Dialog Summarization** is an **advanced Natural Language Processing (NLP)** task that focuses on **distilling essential information** from **multi-turn conversations**. It aims to create **concise summaries** while retaining the **conversational context**. Here are some key aspects:

1. **Input and Output**: The input for Dialog Summarization is typically a conversation or dialogue with multiple participants. The desired output is a brief summary that captures the most important points discussed during the conversation.

2. **Methods**: Dialog Summarization can be achieved through two main approaches:
   - **Abstractive Summarization**: This method involves paraphrasing and rewriting parts of the conversation to create a shorter, coherent summary. It can include information not explicitly present in the original dialogue.
   - **Extractive Summarization**: Extractive summarization selects and rearranges sentences or phrases directly from the dialogue to form the summary, avoiding the generation of new content.

3. **Applications**: Dialog Summarization finds applications in various domains, including:
   - **Customer Support**: Summarizing customer service chats or phone calls to provide quick overviews of issues and resolutions.
   - **Business Meetings**: Creating condensed summaries of spoken or written interactions during meetings and conference calls.
   - **Social Media**: Extracting key points from user discussions, comment sections, or forum threads.
   - **Interviews**: Summarizing interviews conducted by journalists, researchers, or content creators.

4. **Challenges**: Dialog Summarization presents unique challenges, such as handling diverse conversation structures, dealing with informal language, resolving ambiguity, and ensuring the summary is informative while excluding redundant or irrelevant details.

5. **Evaluation**: The quality of Dialog Summarization systems is often assessed using evaluation metrics like ROUGE, which measure the similarity between the generated summary and reference summaries or human-authored summaries. Human judgments are also used for evaluation.

In essence, **Dialog Summarization** is a specialized NLP task that revolves around condensing the **content of conversational dialogues** into **concise summaries**, making it easier for users to **grasp the most critical information** from **discussions** in various **practical contexts**.

### **Zero Shot Inference**

Without providing any **task-specific examples**, we entrust the **model with the task**, expecting it to **autonomously comprehend** the **problem** and **generate the desired outcomes**. It's a test of the **model's inherent understanding and adaptability**.

```python
for _ in range(3):

    dialog, summary = get_random_sample('test')

    prompt = f"""
Summarize the following conversation:

{dialog}

Summary:
    """

    model_completion = model_generation(model=flan_model, prompt=prompt)

    print("\nINPUT PROMPT:")
    print(prompt)
    print("--" * 50)

    print("\nHUMAN BASELINE:")
    print(summary)
    print("--" * 50)

    print("\nFLAN - T5 - Zero Shot:")
    print(model_completion)
    print("--" * 50)
```

Indeed, there's a **noticeable enhancement** in the **model's summarization capabilities**. However, a **pertinent issue persists**: the model appears to be **rearranging the words** from the **input prompt** rather than **providing meaningful context-based summaries**. This limitation, understandably, impacts **the quality of our summarizations**. But there's optimism on the horizon, as there are **avenues for improvement**.

Let's try changing the prompt.

```python
for _ in range(3):

    dialog, summary = get_random_sample('test')

    prompt = f"""
Dialog:

{dialog}

Whats going on?
    """

    model_completion = model_generation(model=flan_model, prompt=prompt)

    print("\nINPUT PROMPT:")
    print(prompt)
    print("--" * 50)

    print("\nHUMAN BASELINE:")
    print(summary)
    print("--" * 50)

    print("\nFLAN - T5 - Zero Shot:")
    print(model_completion)
    print("--" * 50)
```

### **One Shot Inference**

Unlike the **zero shot inference**, where we **do not provide** any example to the model in **one shot inference**, we provide it **1 example**.

```python
for _ in range(2):

    dialog_in, summary_in = get_random_sample('test')
    dialog, summary = get_random_sample('test')

    prompt = f"""
Dialog:
{dialog_in}

What's going on?
{summary_in}

Dialog:
{dialog}

What's going on?
"""

    model_completion = model_generation(model=flan_model, prompt=prompt)

    print("\nINPUT PROMPT:")
    print(prompt)
    print("--" * 50)

    print("\nHUMAN BASELINE:")
    print(summary)
    print("--" * 50)

    print("\nFLAN - T5 - One Shot:")
    print(model_completion)
    print("--" * 50)
```

Not sort of a great improvement let's try few short inference.

### **Few Shot Inference**

```python
def few_shot(n_shots=2, config = None):
    prompt = ''
    dialog, summary = get_random_sample('test')

    for _ in range(n_shots):
        dialog_in, summary_in = get_random_sample('test')
        prompt += f"""
Dialog:
{dialog_in}

What's going on?
{summary_in}
"""

    prompt += f"""
Dialog:
{dialog}

What's going on?
"""

    model_completion = model_generation(model=flan_model, prompt=prompt, config = config)

    print("\nINPUT PROMPT:")
    print(prompt)
    print("--" * 50)

    print("\nHUMAN BASELINE:")
    print(summary)
    print("--" * 50)

    print("\nFLAN - T5 - Few Shot:")
    print(model_completion)
    print("--" * 50)
```

* **max_new_tokens**: This parameter sets a limit on the maximum number of new tokens the language model can generate in its output. It controls the length or size of the generated text, ensuring it doesn't exceed a predefined token count. For example, if you set max_new_tokens to 50, the model will generate a response with a token length of up to 50 tokens, regardless of the input.

* **temperature**: The temperature parameter influences the randomness of the model's output. A higher temperature, like 1.0, results in more diverse and creative responses, sometimes introducing randomness. Conversely, a lower temperature, like 0.2, leads to more deterministic and focused responses, as it reduces randomness and encourages the model to select the most probable next tokens based on its training data.

It's crucial to recognize that **prompt engineering**, while valuable for **certain general purposes**, isn't the **optimal approach** for comprehensive training of **large language models** in **downstream tasks**. **Fine-tuning** is **generally a more effective strategy** for **achieving task-specific performance**. However, when seeking **versatility and quick adaptability**, prompt engineering proves its worth.

Nonetheless, it's essential to understand that the **effectiveness of prompt engineering** is intricately tied to **several factors**. These include the **specific prompt you input**, **the number of samples** providedâ€”whether it's **zero-shot**, **one-shot**, or **few-shot**â€”and the **critical consideration** of staying within the **context window**, which **imposes limits on the total text length.**

Furthermore, **sample diversity** is **a pivotal factor**, especially when dealing with **smaller models**. **Smaller models** often lack the **capacity to effectively utilize prompt engineering** due to their **limited capabilities**. In such cases, **alternative strategies** may be **more appropriate** to achieve desired **results in a more reliable manner**.
